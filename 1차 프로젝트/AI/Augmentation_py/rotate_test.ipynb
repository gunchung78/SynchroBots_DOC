```python
import cv2
import os
import numpy as np

# ----------------------------------------------------------------------
# 1. ì„¤ì • ë³€ìˆ˜
# ----------------------------------------------------------------------

# ì…ë ¥ í´ë”ì™€ ì¶œë ¥ í´ë” ì´ë¦„
INPUT_FOLDER = "Right_Img"
OUTPUT_FOLDER = "Right_Rotate"
# íšŒì „ì‹œí‚¬ ê°ë„ ëª©ë¡ (0ë„ëŠ” ì œì™¸, 10ë„ ê°„ê²©)
ROTATION_ANGLES = [-30, -20, -10, 10, 20, 30] 

# ----------------------------------------------------------------------
# 2. ì´ë¯¸ì§€ íšŒì „ í•¨ìˆ˜ ì •ì˜ (ì´ì „ê³¼ ë™ì¼)
# ----------------------------------------------------------------------

def rotate_image(image, angle):
    """
    ì£¼ì–´ì§„ ì´ë¯¸ì§€ë¥¼ ì§€ì •ëœ ê°ë„(degree)ë§Œí¼ íšŒì „ì‹œí‚µë‹ˆë‹¤.
    ì–‘ìˆ˜ëŠ” ì‹œê³„ ë°˜ëŒ€ ë°©í–¥, ìŒìˆ˜ëŠ” ì‹œê³„ ë°©í–¥ íšŒì „ì…ë‹ˆë‹¤.
    """
    if image is None:
        return None

    # ì´ë¯¸ì§€ ë†’ì´ì™€ ë„ˆë¹„ ê°€ì ¸ì˜¤ê¸°
    (h, w) = image.shape[:2]
    
    # íšŒì „ì˜ ì¤‘ì‹¬ì  ì„¤ì • (ì´ë¯¸ì§€ ì¤‘ì•™)
    center = (w // 2, h // 2)

    # íšŒì „ í–‰ë ¬ (Rotation Matrix) ê³„ì‚°
    M = cv2.getRotationMatrix2D(center, angle, 1.0)

    # ì•„í•€ ë³€í™˜(Affine Transformation) ì ìš©
    rotated_img = cv2.warpAffine(image, M, (w, h))

    return rotated_img

# ----------------------------------------------------------------------
# 3. ë°°ì¹˜ ì²˜ë¦¬ ì‹¤í–‰
# ----------------------------------------------------------------------

def process_rotation_augmentation():
    # ì…ë ¥ í´ë” ì¡´ì¬ ì—¬ë¶€ í™•ì¸
    if not os.path.exists(INPUT_FOLDER):
        print(f"âŒ ì˜¤ë¥˜: ì…ë ¥ í´ë” '{INPUT_FOLDER}'ë¥¼ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤. í´ë”ë¥¼ ìƒì„±í•˜ê³  PNG íŒŒì¼ì„ ë„£ì–´ì£¼ì„¸ìš”.")
        return

    # ì¶œë ¥ í´ë” ìƒì„± (ì—†ìœ¼ë©´ ìƒì„±)
    if not os.path.exists(OUTPUT_FOLDER):
        os.makedirs(OUTPUT_FOLDER)
        print(f"âœ… ì¶œë ¥ í´ë” '{OUTPUT_FOLDER}'ë¥¼ ìƒì„±í–ˆìŠµë‹ˆë‹¤.")

    # ì…ë ¥ í´ë” ë‚´ì˜ ëª¨ë“  PNG íŒŒì¼ ëª©ë¡ ê°€ì ¸ì˜¤ê¸°
    image_files = [f for f in os.listdir(INPUT_FOLDER) if f.lower().endswith('.png')]

    if not image_files:
        print(f"âŒ ì˜¤ë¥˜: '{INPUT_FOLDER}' í´ë”ì— PNG ì´ë¯¸ì§€ê°€ ì—†ìŠµë‹ˆë‹¤.")
        return

    print(f"ğŸ” '{INPUT_FOLDER}'ì—ì„œ {len(image_files)}ê°œì˜ PNG ì´ë¯¸ì§€ë¥¼ ë°œê²¬í–ˆìŠµë‹ˆë‹¤.")
    
    total_generated_images = 0

    for filename in image_files:
        img_path = os.path.join(INPUT_FOLDER, filename)
        
        # ì´ë¯¸ì§€ ì½ê¸°
        original_img = cv2.imread(img_path)

        if original_img is None:
            print(f"âš ï¸ ê²½ê³ : '{filename}' ì´ë¯¸ì§€ë¥¼ ì½ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤. ê±´ë„ˆëœë‹ˆë‹¤.")
            continue

        base_name, ext = os.path.splitext(filename)

        # ì„¤ì •ëœ ê°ë„ë³„ë¡œ íšŒì „ ë° ì €ì¥
        for angle in ROTATION_ANGLES:
            # ì´ë¯¸ì§€ íšŒì „
            rotated_img = rotate_image(original_img, angle)

            # ì €ì¥í•  íŒŒì¼ ì´ë¦„ ìƒì„± (ì˜ˆ: file.png -> file_rotated_-30.png)
            rotated_filename = f"{base_name}_rotated_{angle}{ext}"
            rotated_img_path = os.path.join(OUTPUT_FOLDER, rotated_filename)

            # ì´ë¯¸ì§€ ì €ì¥
            cv2.imwrite(rotated_img_path, rotated_img)
            total_generated_images += 1
            
    print("\n-------------------------------------------------")
    print(f"ğŸ‰ íšŒì „ Augmentation ì™„ë£Œ!")
    print(f"ì›ë³¸ ì´ë¯¸ì§€ ìˆ˜: {len(image_files)}ì¥")
    print(f"ìƒì„±ëœ ê°ë„(Augmentation) ìˆ˜: {len(ROTATION_ANGLES)}ê°œ")
    print(f"ì´ ìƒì„±ëœ íšŒì „ ì´ë¯¸ì§€ ìˆ˜: {total_generated_images}ì¥")
    print(f"ê²°ê³¼ëŠ” '{OUTPUT_FOLDER}' í´ë”ì— ì €ì¥ë˜ì—ˆìŠµë‹ˆë‹¤.")
    print("-------------------------------------------------")

# ìŠ¤í¬ë¦½íŠ¸ ì‹¤í–‰
process_rotation_augmentation()
```
